<center><strong>Contour Mapping of an Oil and Gas Field</strong></center>

Contour maps are an integral part of understanding a reservoir and exploring for possible hidden reserves.  Seismic data provides direction but it is too coarse to provide information at the resolution required to resolve details in the search for hidden reserves in an existing field.  Publicly available existing well records provide a resource for plotting contour maps to further understand the shape and structure of existing fields.


The purpose of this project is to create a map of well locations and a contour map based on the tops reported to the state.  In this project, I selected a field that has mostly been plugged and abandoned. In this field, a reservoir, known as the Dundee.  In the state of Michigan there were many Dundee reservoirs drilled, produced and later plugged, mostly in the 1930's and 1940's. The Dundee is a relatively shallow, porous carbonate that was drilled and produced in the 1930's - 1970's, but fields and single well structures continue to be discovered.  Many of the fields were productive, until it was no longer economical to maintain, and then plugged decades ago.  Some projects have drilled lateral wells through older fields to drain hidden reserves, with various degrees of success.  More successful has been the use of old state records to track down prospects for new vertical wells.

Oil wells in the Dundee require two components to be successful: they require a structural trap, but more often, the more important element is the presence of dolomite.  Dolomite forms when magnesium replaces calcium in the limestone matrix.  The process increases the porosity of the rock.  In this exercise we will look at the structure of an old Dundee field using a contour map and examine some of the wells that were recently drilled and are currently producing in the field.


All information for this project is from publicly available sources.


With respect to python, we will used the numpy meshgrid function and the "tri" function of matplotlib to create and plot triangular elements to generate a contour map.

```python
import matplotlib.pyplot as plt
import matplotlib.tri as tri
import numpy as np
import pandas as pd
```

Data was downloaded from the state of Michigan Department of Environmental Quality to locate:<br>
1) Identify an appropriate field to explore<br>
2) Well locations, depths, well status as well as latitude and longitude<br>
3) Identify geological tops for the Dundee from a separate state database<br>
4) Identify well surface reference altitude<br>
5) Calculate Dundee top subsea depth for each well within the field<br>

The data was assembled into an excel file for use to import into this project.  Below is the code for importing that file.  The file below is available on my Github at: https://github.com/rrbrodeur/Machine-Learning

```python
#Read data:
df = pd.read_excel('FreemanReddingDundeeField.xlsx', header=1, sheet_name='DundeeTops')
```

There are a total of 179 wells.  There are 67 well where the Dundee top calculation results in NAN. Below we will drop the wells that lack the information to calculate the Dundee top.  

```python
df.head()
```

```python
#We will need the longitude, latitude and Dundee top subsea depth as numpy data to use in the "tri" 
#function to triangulate the data to creat a contour map

xx = df.BH_Long
yy = df.BH_Lat
zz = df['Dundee Top Subsea']
```

```python
#I make it a habit to check my data before continuing
print(xx,yy,zz)
```

We see that there are several missing data points that would create a problem with the contouring.  We will create a subset of the data and drop the data points with NaN.  

```python
dfc = df.copy()
dfc.dropna(how='any', inplace=True)
```

```python
x = dfc.BH_Long
y = dfc.BH_Lat
z = dfc['Dundee Top Subsea']
```

We want to see what wells are still producing

```python
prod = df[df.Well_Status=='Producing']
px, py = prod.BH_Long, prod.BH_Lat,
prod
```

We see that there are 21 wells still listed as producing or perhaps were drilled much later in the life of the field.
We're also interested in wells that were drilled as dry holes and wells that are used as salt water disposal wells (SWD) or brine disposal wells.

```python
recent = prod[prod.Permit_Number>50000]
rx, ry = recent.BH_Long, recent.BH_Lat,
recent
```

The date the well was drilled was not part of the data available, so the state permit number was used as a proxy for well drilling date.  Permits are assigned sequentially.
There are only five wells recently drilled (in the past 20 years). 

Next we identify the dry holes and the wells used for salt water disposal (SWD), also known as brine wells:

```python
dry = df[df.Well_Type=='Dry Hole']
dx, dy = dry.BH_Long, dry.BH_Lat,
dry
```

```python
swd = df[df.Well_Type=='Brine Disposal Well']
sx, sy = swd.BH_Long, swd.BH_Lat,
swd
```

There are two dry holes and seven SWD/brine disposal wells in the field.

Let's use a scatter plot to illustrate the locations of the wells and each well type.

```python
fig=plt.figure()
fig.set_size_inches(12,9)
ax=fig.add_axes([0,0,1,1])
ax.scatter(xx,yy,color='k', label = 'All wells')
ax.scatter(x, y, color='gray', label = 'Wells with contour data')
ax.scatter(px, py, color = 'r', label = 'Active Producer')      #producing wells
ax.scatter(sx,sy, marker='^',color='k', s=100, label = 'SWD well')
ax.scatter(dx,dy, marker='X', color='k', s=100, label = 'Dry Hole')
ax.set_xlabel('BH_Long')
ax.set_ylabel('BH_Lat')
ax.set_title('Well Bottom Hole Locations')
plt.legend()
plt.show()
```

```python
dfc.BH_Long.count()
```

We have 112 wells (wells denoted by grey circles in the figure above) that have sufficient information to be used to generate a contour map.  Let's look at the min, max latitude and longitude:

```python
print(dfc.BH_Long.min(), dfc.BH_Long.max(), dfc.BH_Lat.min(), dfc.BH_Lat.max())
```

```python
#We are generating a contour plot of irregularly spaced data
#We define the x & y grid space relative to the longitude and lattitude and generate the grid
ngridx=100
ngridy=100
npts = dfc.BH_Long.count()

xi = np.linspace(dfc.BH_Long.min(), dfc.BH_Long.max(), ngridx)
yi = np.linspace(dfc.BH_Lat.min(), dfc.BH_Lat.max(), ngridy)

# Perform linear interpolation of the data (x,y) on the grid defined by (xi,yi)
triang = tri.Triangulation(x, y)

#Now we interpolate the z componenet 
interpolator = tri.LinearTriInterpolator(triang, z)
Xi, Yi = np.meshgrid(xi, yi)
zi = interpolator(Xi, Yi)
```

```python
#Create contour map and plot wells of interest:
fig, (ax1) = plt.subplots(nrows=1)
ax1.contour(xi, yi, zi, levels=14, linewidths=0.5, colors='k')
cntr1 = ax1.contourf(xi, yi, zi, levels=14, cmap="RdBu_r")
fig.colorbar(cntr1, ax=ax1)
ax1.plot(xx,yy,'o', color='k', ms=6, label = "All Wells") #plot the location of all the wells, even those not used
#ax1.plot(x, y, 'ko', ms=5, label = "Wells with Contour Data")    #ko makes the points an "o", k makes it black, default is blue
ax1.plot(x,y,'o', color='gray', ms=6, label = "Wells Used to Generate Contours")
ax1.plot(px,py,'o', color='orange', ms=6, label = "Producing Wells")
ax1.scatter(sx,sy, marker='^',color='k', s=100, label = 'SWD well')
ax1.plot(rx,ry,'o', color='r', ms=8, label = "Recently Drilled")
#ax1.set(xlim=(-85.070,-85.000), ylim=(43.975, 44.02))
ax1.scatter(dx,dy, marker='X', color='k', s=100, label = 'Dry Hole')
ax1.set(xlim=(dfc.BH_Long.min()-.005, dfc.BH_Long.max()+.005), 
        ylim=(dfc.BH_Lat.min()-.005, dfc.BH_Lat.max()+.005))
ax1.set_title('Dundee Contour of Redding Freeman Field (%d points, %d grid points)' %
              (npts, ngridx * ngridy))
plt.legend()
fig.set_size_inches(18,14)

plt.show()
```

The Dundee is a natural water drive reservoir.  In a water drive reservoir, wells that sit highest in the structure (red contours in the map above) should be the wells that have the longest life and are usually the most prolific oil producers.  Orange circles indicate the wells that are active producers.  Red circles indicate wells that were drilled within the last 20 years. We will zoom in on the region of the more recently drilled wells (north west portion of the field).

```python
#Zoom in on contour map and plot wells of interest:
fig, (ax1) = plt.subplots(nrows=1)
ax1.contour(xi, yi, zi, levels=14, linewidths=0.5, colors='k')
cntr1 = ax1.contourf(xi, yi, zi, levels=14, cmap="RdBu_r")
fig.colorbar(cntr1, ax=ax1)
ax1.plot(xx,yy,'o', color='k', ms=6, label = "All Wells") #plot the location of all the wells, even those not used
#ax1.plot(x, y, 'ko', ms=5, label = "Wells with Contour Data")    #ko makes the points an "o", k makes it black, default is blue
ax1.plot(x,y,'o', color='gray', ms=6, label = "Wells Used to Generate Contours")
ax1.plot(px,py,'o', color='orange', ms=8, label = "Producing Wells")
ax1.scatter(sx,sy, marker='^',color='k', s=150, label = 'SWD well')
ax1.plot(rx,ry,'o', color='r', ms=8, label = "Recently Drilled Producing Wells")
#ax1.set(xlim=(-85.070,-85.000), ylim=(43.975, 44.02))
ax1.scatter(dx,dy, marker='X', color='k', s=150, label = 'Dry Hole')
ax1.set(xlim=(dfc.BH_Long.min()-.005, -85.03), 
        ylim=(43.99, dfc.BH_Lat.max()+.005))
ax1.set_title('Zoomed View of Dundee Contour of Redding Freeman Field')
plt.legend()
fig.set_size_inches(18,14)

plt.show()
```

In the above contour map we see four "peaks" (contours becoming increasingly red in color) and two "valleys" (contours becoming increasingly blue in color). The actively producing wells most recently drilled (red circles) are located in two of the valleys and only one of the peaks. Why would an actively producing well be drilled in the deepest part of the reservoir? That makes no sense based on the contours of the reservoir. However, examining the logs of that well (that analysis is not part of this presentation), the well is dolomitized throughout the Dundee region. Another important item to note is that the surrounding wells were all drilled in the 1930's. In that era, geological tops were identified from drill cuttings as they were brought up to the surface. In both cases the newly drilled well literally defines the valley of the contour map. My interpretation is that the surrounding wells have an inherent bias in their reported depths.  Unfortunately, the data required to reassess the actual depths and formation tops has been lost for most of the wells in this field.  The more recently drilled wells would likely represent a more accurate representation of the true depth.  

The Dundee has been a prolific oil producing formation and there are cases where wells drilled over 50 years ago continue to produce today.  Predicting oil from structure is an important variable in the development of a reservoir and contour mapping can provide insight on where to search for missed reserves.  However, replacement of the limestone with dolomite is an equally important variable, as it provides the porosity for the oil to saturate the rock.  Unfortunately the presence of dolomite can only be determined after the well is drilled.  Seismic information helps define the structure but it cannot determine the presence or absence of dolomite.
